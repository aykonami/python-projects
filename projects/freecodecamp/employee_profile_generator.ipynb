```python
"""
project: employee profile generator
description: generates employee details and extracts information from employee code
"""

# basic employee details
first_name = 'John'
last_name = 'Doe'

# create full name
full_name = first_name + ' ' + last_name

# address information
address = '123 Main Street'
address += ', Apartment 4B'

# age and basic info
employee_age = 28
employee_info = full_name + ' is ' + str(employee_age) + ' years old'
print(employee_info)

# experience details
experience_years = 5
experience_info = 'Experience: ' + str(experience_years) + ' years'
print(experience_info)

# job details
position = 'Data Analyst'
salary = 75000

# formatted employee card
employee_card = f'Employee: {full_name} | Age: {employee_age} | Position: {position} | Salary: ${salary}'
print(employee_card)

# employee code analysis
employee_code = 'DEV-2026-JD-001'

# extract department
department = employee_code[0:3]
print(department)

# extract year and initials
year_code = employee_code[4:8]
initials = employee_code[9:11]

print(year_code)
print(initials)

# extract last three digits
last_three = employee_code[-3:]
print(last_three)
```
