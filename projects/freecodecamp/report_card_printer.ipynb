```python
"""
project: report card printer
description: demonstrates basic python data types and type checking
"""

# student name
name = 'Alice'
print(name, type(name))

# student status
is_student = True
print(is_student, type(is_student))

# student age
age = 20
print(age, type(age))

# student score
score = 80.5

# check data type of score
print(isinstance(score, int))
print(isinstance(score, float))

# display score and its type
print(score, type(score))
```
