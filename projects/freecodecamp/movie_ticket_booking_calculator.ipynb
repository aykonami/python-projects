```python
"""
project: movie ticket booking calculator
description: checks eligibility, applies discount, and calculates final ticket price
"""

# base details
base_price = 15
age = 21
seat_type = 'Gold'
show_time = 'Evening'

# eligibility checks
if age > 17:
    print('user is eligible to book a ticket')

if age >= 21:
    print('user is eligible for evening shows')
else:
    print('user is not eligible for evening shows')

# membership discount
is_member = True
is_weekend = False
discount = 0

if is_member:
    discount = 3
    print('user qualifies for membership discount')
else:
    print('user does not qualify for membership discount')

print('discount:', discount)

# combined condition discount
if is_member and age >= 21:
    discount = 3
    print('user qualifies for membership discount')
else:
    print('user does not qualify for membership discount')

print('discount:', discount)

# reset and test again
is_member = False
is_weekend = False
discount = 0

if is_member and age >= 21:
    discount = 3
    print('user qualifies for membership discount')
else:
    print('user does not qualify for membership discount')

print('discount:', discount)

# weekend charges
extra_charges = 0

if is_weekend:
    extra_charges = 2
    print('extra charges will be applied')
else:
    print('no extra charges will be applied')

print('extra charges:', extra_charges)

# weekend or evening charges
extra_charges = 0

if is_weekend or show_time == 'Evening':
    extra_charges = 2
    print('extra charges will be applied')
else:
    print('no extra charges will be applied')

print('extra charges:', extra_charges)

# booking conditions
if age >= 21:
    print('ticket booking condition satisfied')
else:
    print("ticket booking failed due to restrictions")

if age >= 21 or age >= 18 and show_time != 'Evening':
    print('ticket booking condition satisfied')
else:
    print('ticket booking failed due to restrictions')

if age >= 21 or age >= 18 and (show_time != 'Evening' or is_member):
    print('ticket booking condition satisfied')

# service charges
if age >= 21 or age >= 18 and (show_time != 'Evening' or is_member):
    print('ticket booking condition satisfied')

    service_charges = 0

    if seat_type == 'Premium':
        service_charges = 5
    elif seat_type == 'Gold':
        service_charges = 3
    else:
        service_charges = 1

    print('service charges:', service_charges)

    # final price calculation
    final_price = base_price + extra_charges + service_charges - discount
    print('final price of ticket:', final_price)

else:
    print('ticket booking failed due to restrictions')
```
