```python
"""
project: bill splitter
description: calculates total bill, adds tip, and splits cost among friends
"""

# number of people splitting the bill
num_of_friends = 4

# item costs
appetizers = 37.89
main_courses = 57.34
desserts = 39.39
drinks = 64.21

# calculate total bill
running_total = appetizers + main_courses + desserts + drinks
print('total bill so far:', running_total)

# calculate tip (25%)
tip = running_total * 0.25
print('tip amount:', tip)

# add tip to total
running_total += tip
print('total with tip:', running_total)

# split bill among friends
each_person = running_total / num_of_friends
print('bill per person:', each_person)

# final rounded result
each_pays = round(each_person, 2)
print('each person pays:', each_pays)
```
