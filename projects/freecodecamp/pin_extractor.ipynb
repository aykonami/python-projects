```python
"""
project: pin extractor
description: extracts numeric codes from poems based on word lengths and line positions
"""

def pin_extractor(poems):
    secret_codes = []

    # loop through each poem
    for poem in poems:
        secret_code = ""

        # split poem into lines
        lines = poem.split("\n")

        # process each line
        for line_index, line in enumerate(lines):
            words = line.split()

            # check if index exists in words
            if len(words) > line_index:
                # add length of the selected word
                secret_code += str(len(words[line_index]))
            else:
                # add 0 if no word exists
                secret_code += "0"

        secret_codes.append(secret_code)

    return secret_codes


# sample poems
poem = """Stars and the moon
shine in the sky
white and
until the end of the night"""

poem2 = """The grass is green\nhere and there\nhoping for rain\nbefore it turns yellow"""

poem3 = """There\nonce\nwas\na\ndragon"""

# run function
print(pin_extractor([poem, poem2, poem3]))
```
